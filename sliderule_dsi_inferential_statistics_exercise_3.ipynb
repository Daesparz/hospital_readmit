# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

From the scatter plot, we observe that the rate of readmissions decrease slightly with the increment of discharges in hospitals. Additionaly, we can see that the density of data concentrates on discharges less than 1000. However, there are not clear trendings in this graph to extract conclusions itself. Drawing **linear regressions with 50 paired bootstraps replicate**, we got:

```python
def draw_bs_pairs_linreg(x, y, size=1):
    
    inds = np.arange(len(x))
    bs_slope = np.empty(size)
    bs_intercept = np.empty(size)
    
    for i in range(size):
        bs_inds = np.random.choice(inds, size = len(inds))
        bs_x = [x[j] for j in bs_inds]
        bs_y = [y[j] for j in bs_inds]
        bs_slope[i], bs_intercept[i] = np.polyfit(bs_x , bs_y, 1)
    return bs_slope, bs_intercept        
```

```python
bs_slope, bs_intercept = draw_bs_pairs_linreg(x, y, 50)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

xi = np.array([x[0], x[-1]])
for i in range(50):  
    yi = bs_slope[i]*xi + bs_intercept[i]
    ax.plot(xi, yi, color='xkcd:aquamarine', linewidth=0.1)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

Effectively, the **relation between rate of admissions and discharges is negative**. Regressions consist in soft decreasing slopes, as we can see in the following histogram of the drawn slopes:

We apply *z score criteria* to deal with outliers:

```python
def delete_outliers(x, y):
    z = np.abs(stats.zscore(y))
    threshold = 3
    inds = np.where(np.abs(z)<= 3)
    return [x[i] for i in inds[0]], [y[i] for i in inds[0]]
```

```python
xa_clean, ya_clean  = delete_outliers(xa, ya)
xb_clean, yb_clean = delete_outliers(xb, yb)
xc_clean, yc_clean = delete_outliers(xc, yc)
```

Then, it inspected **how normal looks readmission ratio distributions**. Defining a ecdf function, it's possible to graph the **cumulative probability in each group** and compare each one with his simulated normal distribution (using the mean and standard deviation of the group as *mu* and *sigma* of the normal distribution)

```python
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1)/n
    return x, y 
```

```python
ai, aj = ecdf(ya_clean)
bi, bj = ecdf(yb_clean)
ci, cj = ecdf(yc_clean)

ai_the, aj_the = ecdf(np.random.normal(np.mean(ya_clean),np.std(ya_clean),len(ya_clean)))
bi_the, bj_the = ecdf(np.random.normal(np.mean(yb_clean),np.std(yb_clean),len(yb_clean)))
ci_the, cj_the = ecdf(np.random.normal(np.mean(yc_clean),np.std(yc_clean),len(yc_clean)))
```

After to graph the ECDF of each group and note the similarity with the normal version of every cluster, we proceed to do the **t-statistic tests**. We have three groups and we can apply t-statistic between two, so we repeat the test three times. Additionally, we inspect results of an **ANOVA test**, that allows comparing all groups at the same time.

*Note: t-statistic and ANOVA tests are parametric tests. We are using them under the premise that clusters represent normal distributions.*

**t-statisctis test**

```python
ttest,p1 = stats.ttest_ind(ya_clean, yb_clean)
print ('t-statistics of two sampled = %.5f, p = %.10f' %(ttest, p1))
```

```python
ttest,p2 = stats.ttest_ind(yb_clean, yc_clean)
print ('t-statistics of two sampled = %.5f, p = %.10f' %(ttest, p2))
```

```python
ttest,p3 = stats.ttest_ind(ya_clean, yc_clean)
print ('t-statistics of two sampled = %.5f, p = %.10f' %(ttest, p3))
```

In all cases, considering a significance level of 0.01, we reject the null hypothesis, because *p << 0.01*. Then, there is a significant difference in groups and we can’t treat them as the same.


**ANOVA test**

```python
F, p = stats.f_oneway(ya_clean, yb_clean, yc_clean)
F, p
```

Reinforcing the prior result, we can conclude that statistical tests suggest that we should **reject the null hypothesis**.

We got statistical significance to say that the three analyzed groups don't have the same distribution, then we can not establish the same statistics values as quality standards for all spectral of data.

However, specific analysis per groups **could indicate that there is no practical significance** in those results. According to means and standard deviation calculated, metrics are similar and the rejection of the null hypothesis could be the effect of the size of the data. Computing **Cohen's d**:

```python
sp = np.sqrt(((len(ya_clean)-1)*np.var(ya_clean)+(len(yb_clean)-1)*np.var(yb_clean))/(len(ya_clean)+len(yb_clean)-1))
d = (np.mean(ya_clean)-np.mean(yb_clean))/sp

print ("Cohen'd = %.3f"%(d))
```

In the first case, we rejected the null hypothesis comparing group 1 and 2 because p was << 0.01, so these results are statistically significant. There is evidence that distributions are different, although they have similar means. However, the **difference is not practically significant** (Cohen's *d* coefficient less than 0.2 is interpreted as the difference between the two groups have a little or no effect)

```python
sp = np.sqrt(((len(ya_clean)-1)*np.var(ya_clean)+(len(yc_clean)-1)*np.var(yc_clean))/(len(ya_clean)+len(yc_clean)-1))
d = (np.mean(ya_clean)-np.mean(yc_clean))/sp

print ("Cohen'd = %.3f"%(d))
```

```python
sp = np.sqrt(((len(yb_clean)-1)*np.var(yb_clean)+(len(yc_clean)-1)*np.var(yc_clean))/(len(yb_clean)+len(yc_clean)-1))
d = (np.mean(yb_clean)-np.mean(yc_clean))/sp

print ("Cohen'd = %.3f"%(d))
```

In the second case (Group 1 and Group 3), interpretation is a **medium effect** and in the third case (Group 2 and Group 3), is a **small effect**. We didn't find a large effect in any case; in other words, practical significance says that difference could not be meaningful in real life. **The recommendation is to get more data and details of every analyzed hospital before determining new Medicare payments regulatory policies.**
